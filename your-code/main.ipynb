```python
#Returns the last year of commit activity grouped by week. The days array is a group of commits per day, starting on Sunday.

url = 'https://api.github.com/repos/ta-data-lis/lab-mysql-first-queries/stats/commit_activity'
repo=requests.get(url,headers=headers)
print(repo.status_code)
repo.json()
commit_activity = pd.DataFrame(repo.json())

#Conver unix timestamp string to datetime
commit_activity['converted']= pd.to_datetime(commit_activity['week'], unit='s')
#There weren't any commits in the las two weeks as seen bellow in commit_activity


#Clean data
commit_activity =commit_activity.rename(columns={'total': 'Total_Commits', 'converted': 'Day'})
commit_activity =commit_activity.drop(['week', 'days'],  axis=1)
commit_activity =commit_activity[['Day', 'Total_Commits']]

#Total_Commits in the last two weeks
commit_activity.tail(2)
```
